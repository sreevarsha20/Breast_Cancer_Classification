Importing the Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
# --------------------- Logistic Regression ---------------------
lr_model = LogisticRegression(max_iter=10000)
lr_model.fit(X_train, Y_train)
lr_train_accuracy = accuracy_score(Y_train, lr_model.predict(X_train))
lr_test_accuracy = accuracy_score(Y_test, lr_model.predict(X_test))
```

SVM Model

```python
# --------------------- SVM Model ---------------------
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, Y_train)
svm_train_accuracy = accuracy_score(Y_train, svm_model.predict(X_train))
svm_test_accuracy = accuracy_score(Y_test, svm_model.predict(X_test))
```

Random Forest Model

```python
# --------------------- Random Forest Model ---------------------
rf_model = RandomForestClassifier()
rf_model.fit(X_train, Y_train)
rf_train_accuracy = accuracy_score(Y_train, rf_model.predict(X_train))
rf_test_accuracy = accuracy_score(Y_test, rf_model.predict(X_test))
```

Model Comparision

```python
# --------------------- Model Comparison ---------------------
import pandas as pd

comparison_df = pd.DataFrame({
    'Model': ['Logistic Regression', 'SVM', 'Random Forest'],
    'Training Accuracy': [lr_train_accuracy, svm_train_accuracy, rf_train_accuracy],
    'Test Accuracy': [lr_test_accuracy, svm_test_accuracy, rf_test_accuracy]
})

print("\n--- Model Performance Comparison ---")
print(comparison_df)
```

Plotting Accuracy

```python
models = comparison_df['Model']
train_scores = comparison_df['Training Accuracy']
test_scores = comparison_df['Test Accuracy']

x = range(len(models))
plt.figure(figsize=(8, 5))
plt.bar(x, train_scores, width=0.4, label='Train Accuracy', align='center')
plt.bar([i + 0.4 for i in x], test_scores, width=0.4, label='Test Accuracy', align='center')
plt.xticks([i + 0.2 for i in x], models)
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.legend()
plt.show()
```

```python
!pip install gradio --quiet
import gradio as gr
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC

# Load and prepare the dataset
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
X = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
Y = breast_cancer_dataset.target

# Train the model using SVC (Support Vector Classifier)
svm_model = SVC(kernel='linear')
svm_model.fit(X, Y)

# Prediction function
def predict_cancer(*inputs):
    input_data = np.array(inputs).reshape(1, -1)
    prediction = svm_model.predict(input_data)
    return "Malignant" if prediction[0] == 0 else "Benign"

# Creating sliders for each feature
inputs = []
for feature in breast_cancer_dataset.feature_names:
    feature_data = X[feature]
    slider = gr.Slider(
        minimum=float(feature_data.min()),
        maximum=float(feature_data.max()),
        step=0.01,
        value=float(feature_data.mean()),
        label=feature
    )
    inputs.append(slider)

# Launching Gradio Interface
demo = gr.Interface(
    fn=predict_cancer,
    inputs=inputs,
    outputs=gr.Text(label="Prediction"),
    title="Breast Cancer Predictor",
    description="Move the sliders to set values for the 30 features and get a prediction: Malignant or Benign"
)

demo.launch()
```
